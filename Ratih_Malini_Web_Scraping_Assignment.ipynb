```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
#Mengakses Web

url = 'https://id.theasianparent.com/'
response = requests.get(url)
```

```python
#Melakukan parsing

soup = BeautifulSoup(response.text, 'html.parser')
```

```python
#Inspect elemen yang dituju

article = soup.select('.ui-list-thumb li')
```

```python
#Mengambil data dari elemen ui-list-thumb --> artikel terpopuler

article
```

```python
#Menghitung banyak artikel terpopuler

len(article)
```

```python
#Mengambil 1 contoh artikel yang pertama

article[0]
```

```python
#Mengambil data title dan url

results = []
for item in article:
  data = dict()

  #data collection
  data['title']= item.select_one('.ui-list-thumb a').get('aria-label')
  data['url']= item.select_one('.ui-list-thumb a').get('href')
  results.append(data)
```

```python
#Menampilkan data title dan url dari artikel terpopuler

df_bs4 = pd.DataFrame(results)
df_bs4
```
